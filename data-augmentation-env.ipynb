```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import shutil
%matplotlib inline
```

```python
# This function will plot images in the form of a grid with 1 row and 10 columns where images are placed
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1,
        height_shift_range=0.1, shear_range=0.15, zoom_range=0.1,
        channel_shift_range=10., horizontal_flip=True)
```

```python
# Change the working directory to '/kaggle/working'
os.chdir('/kaggle/working')

# Define the source directory in the input path
source_directory = '/kaggle/input/sample-dataset-2/dogs-vs-cats'

if os.path.isdir('my_dataset') is False:
    # Copy the entire source directory to the working directory
    shutil.copytree(source_directory, 'my_dataset')
```

```python
chosen_image = random.choice(os.listdir('/kaggle/working/my_dataset/train/dog'))
```

```python
image_path = '/kaggle/working/my_dataset/train/dog/' + chosen_image
```

```python
if os.path.isfile(image_path):
    print(f"The file at '{image_path}' exists.")
else:
    print(f"The file at '{image_path}' does not exist.")
```

```python
assert os.path.isfile(image_path)
```

```python
# Obtain image
image = np.expand_dims(plt.imread(image_path), 0)
plt.imshow(image[0])
```

```python
# Generate batches of augmented images from this image
aug_iter = gen.flow(image)
```

```python
# Get 10 samples of augmented images
aug_images = [next(aug_iter) [0].astype(np.uint8) for i in range(10)]
```

```python
# Augmented images
plotImages(aug_images)
```
